```python
import socket
import struct
import numpy as np
import cv2
```

```python
def main(server_ip):
    server_address = (server_ip, 65432)
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.connect(server_address)

    try:
        while True:
            # Receive frame size
            frame_size_data = sock.recv(4)
            if not frame_size_data:
                break
            frame_size = struct.unpack('!I', frame_size_data)[0]

            # Receive frame data
            frame_data = b''
            while len(frame_data) < frame_size:
                packet = sock.recv(frame_size - len(frame_data))
                if not packet:
                    break
                frame_data += packet

            #print("Frame_data", frame_data)

            yuv_frame = np.frombuffer(frame_data, dtype=np.uint8).reshape((frame_height, frame_width * 2))

            # Convert YUV422 to BGR
            frame = cv2.cvtColor(yuv_frame, cv2.COLOR_YUV2BGR_YV12)

            # Display the frame
            cv2.imshow('Frame', frame)

            # Receive byte array
            byte_array_size = 10 * 1 + 10 * 2  # 10 INT8 + 10 Float16
            byte_array_data = sock.recv(byte_array_size)
            byte_array = struct.unpack('!10b10e', byte_array_data)
            int8_values = byte_array[:10]
            float16_values = byte_array[10:]

            #print("INT8 values:", int8_values)
            #print("Float16 values:", float16_values)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    finally:
        sock.close()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    # Define the frame height and width
    frame_height, frame_width = 360, 640  # Example resolution, should match server-side settings
    main('127.0.0.1')
```
